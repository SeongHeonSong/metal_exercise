1. design.get_units() returns default unit 'mm'
2. design.parse_value('any number with any unit') returns ONLY the value 
converted to default unit 'mm'.
3. design.parse_value() is a function which converts any number with any unit 
in any from of data structure(single value representation, list, dictionary, ...)

```python
print('Design default units for length: ', design.get_units())
print('\n')
print('Example 250 micrometer parsed to design units: ', design.parse_value('0.250 um'), design.get_units())
print('\n')

dictionary={'key_in_cm':'1.2 cm', 
            'key_in um': '50 um'}
print('Example parse dict: ',design.parse_value(dictionary))
print('\n')

a_list=['1 m', '1 mm', '1 um', '1 nm']
print('Example parse list: ', design.parse_value(a_list))
print('\n')

print(design.parse_value('2*2 um'), 
      design.parse_value('2 um + 5 um'))
```
